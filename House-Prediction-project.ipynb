```python
from sklearn.compose import ColumnTransformer, make_column_transformer
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder,OrdinalEncoder


# define a pipe line with feature preprocessing including Imputer for filling Nan values and Ordinalencoder and onehotencoder

def create_transformer(cat_features,num_features,ord_features,data):

    num_transformer = Pipeline(
        steps = [
            ('imputer', SimpleImputer(strategy='mean')),

        ]
    )



    cat_transformer = Pipeline(
        steps = [
            ('imputer', SimpleImputer(strategy='most_frequent')),
            ('onehot', OneHotEncoder(sparse=False, handle_unknown='ignore')),

        ]
    )


    ord_transformer = Pipeline(steps = [('imputer', SimpleImputer(strategy = 'most_frequent')),
                                           (('odi', (OrdinalEncoder())))])



    preprocessor = ColumnTransformer(
        transformers = [
            ('num', num_transformer, num_features),
            ('cat', cat_transformer, cat_features),
            ('ord', ord_transformer, ord_features),
        ]
    )
    data = preprocessor.fit_transform(data)
    return data


```

```python
X=create_transformer(cat_features,num_features,ord_features,train_df)
X.shape
```

```python
from sklearn.model_selection import train_test_split
# Split the data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30)
```

```python
from sklearn.metrics import mean_squared_error,r2_score
from sklearn.metrics import accuracy_score
from sklearn.metrics import mean_absolute_error


def evaluate_model(dt_classifier,X_train,y_train,X_test,y_test):

    
    y_pre=dt_classifier.predict(X_train)
   
    print("Train MSE:",mean_squared_error(y_train,y_pre))
    print("Train MAE:",mean_absolute_error(y_train,y_pre))
    print("Train Root MSE:",mean_squared_error(y_train,y_pre,squared=False))
    print("Train R2:",r2_score(y_train,y_pre,multioutput='variance_weighted'))
    
  
    #results on test
    print("*************")
    y_pre_test=dt_classifier.predict(X_test)
    
    print("Test MSE:",mean_squared_error(y_test,y_pre_test))
    print("Test MAE:",mean_absolute_error(y_test, y_pre_test))
    print("Test Root MSE:",mean_squared_error(y_test, y_pre_test,squared=False))
    print("Test R2:",r2_score(y_test, y_pre_test,multioutput='variance_weighted'))
    
   
```
